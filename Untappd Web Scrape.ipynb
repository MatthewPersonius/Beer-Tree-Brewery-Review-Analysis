```python
data['IBU'] = data['IBU'].str[:-3]
```

```python
data['Rating'] = data['Rating'].str[1:]
```

```python
data.to_csv("All_Beer_Tree_Untappd.csv")
```

```python
missing_info=data[data.Rating  == "N/A)"]
```

```python
missing_info
```

```python
cleaned_df=data[data.Rating  != "N/A)"]
```

```python
cleaned_df
```

```python
cleaned_df['Beer Type'].unique()
```

```python
cleaned_df["IPA"] = cleaned_df['Beer Type'].str.contains('IPA')
```

```python
cleaned_df
```

```python
cleaned_df.to_csv("No_Missing_Beer_Tree_Untappd.csv")
```

```python
only_IPA=cleaned_df[cleaned_df.IPA == True]
```

```python
only_IPA
```

```python
only_IPA["Rating"] = pd.to_numeric(only_IPA["Rating"])
only_IPA["ABV%"] = pd.to_numeric(only_IPA["ABV%"])
only_IPA["Number of ratings"] = (only_IPA["Number of ratings"].str.split()).apply(lambda x: float(x[0].replace(',', '')))
```

```python
only_IPA["Number of ratings"] = pd.to_numeric(only_IPA["Number of ratings"])
```

```python
only_IPA.dtypes
```

```python
only_IPA.to_csv("IPA_Beer_Tree_Untappd.csv")
```
