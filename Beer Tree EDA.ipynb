```python
no_na=pd.read_csv("No_Missing_Beer_Tree_Untappd.csv")
```

```python
no_na=no_na.sort_values(by="Rating", ascending=False)
```

```python
no_na
```

```python
no_na=no_na[no_na["ABV%"].str.contains("N/")==False]
```

```python
no_na["Rating"] = pd.to_numeric(no_na["Rating"])
no_na["ABV%"] = pd.to_numeric(no_na["ABV%"])
no_na["Number of ratings"] = (no_na["Number of ratings"].str.split()).apply(lambda x: float(x[0].replace(',', '')))
#no_na["Number of ratings"] = pd.to_numeric(no_na["Number of ratings"])
```

```python
no_na.dtypes
```

```python
plt.scatter(x=no_na['ABV%'], y=no_na['Rating'], alpha=0.3)
plt.xlabel("Alcohol Percentage")
plt.ylabel("Rating")
plt.title("Alcohol Percentage vs. Ratings ")
```

```python

plt.scatter(x=thousand['ABV%'], y=thousand['Rating'], alpha=0.4)
plt.xlabel("Alcohol Percentage")
plt.ylabel("Rating")
z = np.polyfit(thousand['ABV%'], thousand['Rating'], 1)
p = np.poly1d(z)
plt.plot(thousand['ABV%'],p(thousand['ABV%']),"r--")

plt.title("Alcohol Percentage vs. Ratings for all beers with over 1000 reviews ")
```
